#ИМПОРТИРУЕМ ДАННЫЕ ДЛЯ ЭТАПА РЕЛАКСАЦИИ

```python
data_relax001 = pd.read_csv("Etap2t001.txt", sep="\t")
data_relax001 = data_relax001.drop(columns=[data_relax001.columns[0], data_relax001.columns[8]])

data_relax = pd.read_csv("Etap2.txt", sep="\t")
data_relax = data_relax.drop(columns=[data_relax.columns[0], data_relax.columns[8]])
```

```python
data_relax001.head(4)
```

```python
data_relax.head(4)
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.plot(data_relax001["Step"], data_relax001['Enthalpy'], label="T=0.01K")
plt.plot(data_relax["Step"], data_relax['Enthalpy'], label="T=300K")
plt.title("График зависимости Энтальпии от Шага")
plt.ylabel("Энтальпия, [эВ]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(data_relax001["Step"], data_relax001['Enthalpy'], label="T=0.01K")
plt.scatter(data_relax["Step"], data_relax['Enthalpy'], label="T=300K",marker='s')
plt.title("График зависимости Энтальпии от Шага")
plt.ylabel("Энтальпия, [эВ]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.savefig("ЭнтальпияОтШага.png")
```

```python
plt.figure(figsize=(12, 7))
plt.plot(data_relax001["Step"], data_relax001['PotEng'], label="T=0.01K", linewidth=5, linestyle="--")
plt.plot(data_relax["Step"], data_relax['PotEng'], label="T=300K", linewidth=1)
plt.title("График зависимости Потенциальной Энергии от Шага")
plt.ylabel("Потенциальная энергия, [эВ]")
plt.xlabel("Шаг")
plt.savefig("ПотенциальнаяЭн.png")
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.plot(data_relax001["Step"], data_relax001['Press'], label="T=0.01K")
plt.plot(data_relax["Step"], data_relax['Press'], label="T=300K")
plt.title("График зависимости Давления от Шага")
plt.ylabel("Давление[Бар]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(data_relax001["Step"], data_relax001['Press'], label="T=0.01K")
plt.scatter(data_relax["Step"], data_relax['Press'], label="T=300K",marker='s')
plt.title("График зависимости Давления от Шага")
plt.ylabel("Давление[Бар]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.savefig("ГрафикДав.png")
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(data_relax001["Step"], data_relax001['Press'], label="T=0.01K")
plt.plot(data_relax["Step"], data_relax['Press'], label="T=300K")
plt.title("График зависимости Давления от Шага")
plt.grid(True)
plt.ylabel("Давление")
plt.xlabel("Шаг")
plt.ylim(-10, 200)
plt.legend()

plt.subplot(2, 2, 2)
plt.scatter(data_relax001["Step"], data_relax001['Press'], label="T=0.01K")
plt.scatter(data_relax["Step"], data_relax['Press'], label="T=300K",marker='s')
plt.title("График зависимости Давления от Шага")
plt.grid(True)
plt.ylabel("Давление")
plt.xlabel("Шаг")
plt.ylim(-10, 200)
plt.legend()
```

#ИСПОРТИРУЕМ ДАННЫЕ ДЛЯ ЭТАПА ПЛАВЛЕНИЯ

```python
data_relax3300 = pd.read_csv("Etap3t300.txt", sep="\t")
data_relax3300 = data_relax3300.drop(columns=[data_relax3300.columns[0], data_relax3300.columns[8]])
```

```python
data_relax3300.head(4)
```

```python
plt.figure(figsize=(20, 10))
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
plt.subplot(2, 2, 1)
plt.plot(data_relax3300["Step"], data_relax3300['Enthalpy'])
plt.ylabel("Энтальпия, [эВ]")
plt.xlabel("Шаг")
plt.title("График зависимости Энтальпии от Шага")
plt.vlines(65000, -45000, -10000, linestyles="dashed", color="orange")
plt.grid(True)
plt.legend()


plt.subplot(2, 2, 2)
plt.plot(data_relax3300["Step"], data_relax3300['Press'])
plt.vlines(65000, 0, 175000, linestyles="dashed", color="orange")
plt.title("График зависимости Давления от Шага")
plt.ylabel("Давление, [Бар]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(data_relax3300["Step"], data_relax3300['Temp'])
plt.title("График зависимости Температуры от Шага")
plt.ylabel("Температура, [К]")
plt.xlabel("Шаг")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(data_relax3300["Step"], data_relax3300['PotEng'])
plt.title("График зависимости Потенциальной Энергии от Шага")
plt.ylabel("Пот.Энергия, [эВ]")
plt.xlabel("Шаг")
plt.vlines(65000, -46000, -39000, linestyles="dashed", color="orange")
plt.grid(True)
plt.legend()

plt.savefig("Плавление.png")
```

```python
plt.figure(figsize=(15, 10))
plt.title("График зависимости Радиальной функции распределения от РадиусаОбрезки")
plt.ylabel("g(r)")
plt.xlabel("Парный радиус[А]")
plt.grid()
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
#plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
#plt.style.use("bmh")

#plt.subplot(5, 2, 1)

for i in range(0, 11, 2):
    link = "/content/0plav." + str(i) + ".txt"
    plavi = pd.read_csv(link,  delim_whitespace=True, header=None, names=["PairSepDist", "g(r)"])
    plavi.head(3)
    plt.plot(plavi.PairSepDist, plavi["g(r)"], label=f"Шаг номер {i* 10000}")

plt.vlines(6.3, 0, 9, linestyle="dashed", label="НачалоПлавления", color="black")
plt.legend()
plt.savefig("РадиалФРфулл")
```
